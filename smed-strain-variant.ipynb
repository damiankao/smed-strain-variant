## Analysis of differences among S. mediterranea strains from different labs

### Data used

Available S. mediterranea raw data was downloaded from NCBI and organized (`ngsData-download.ipynb`). An index file (`data.info`) was made to describe each `.fastq` file. 

```python
import sys, os
from collections import Counter
#import custom functions for displaying tables, bash commands
sys.path.append(os.path.abspath("/home/damian/"))
from dk_ipython import *

dataInfo = [line.strip().split('\t') for line in open('/home/share/ngsData/smed/data.info')]
fastqs = Counter([x.strip().split('_')[0] for x in os.listdir('/home/share/ngsData/smed/fastq')])
illum = dict([(x[0],x) for x in dataInfo if x[9].lower().find('illumina') != -1])

print 'Number of illumina sequenced libraries:', len(illum)
print 'Number of labs:', len(set([x[2] for x in dataInfo]))
print 'Labs:'
print '\n'.join(set(['  ' + x[2] for x in dataInfo]))
```

### Trimming reads

All Illumina reads were trimmed with Trimmomatic. Minimum length set at 30.

```python
#generate trimmmomatic commands and run with:
#parallel :::: trimm.cmd > trimm.out 2> trimm.err 
trimCmdSEA = 'java -jar /home/share/software/Trimmomatic-0.32/trimmomatic-0.32.jar SE -threads 1 '
trimCmdPEA = 'java -jar /home/share/software/Trimmomatic-0.32/trimmomatic-0.32.jar PE -threads 1 '
trimCmdB = ' LEADING:20 TRAILING:20 MINLEN:30 ILLUMINACLIP:/home/share/software/Trimmomatic-0.32/adapters/adapters.fa:2:30:12:1:false TOPHRED33'

trimmomaticBash = ''
for lib, count in fastqs.items():
    if lib in illum:
        fulllib = '/home/share/ngsData/smed/fastq/' + lib
        phred = detectPhred(fulllib + '_1.fastq.gz', 5000) #in dk_ipython
        if phred == 'Sanger':
            phred = '-phred33 '
        else:
            phred = '-phred64 '
        if count == 1:
            #single end
            trimmomaticBash += trimCmdSEA + phred + fulllib + '_1.fastq.gz ' + lib + '_1.trimmed.fastq.gz ' + trimCmdB + '\n'
        else:
            #paired end
            trimmomaticBash += trimCmdPEA + phred + fulllib + '_1.fastq.gz ' + fulllib + '_2.fastq.gz ' + \
            lib + '_1.paired.trimmed.fastq.gz ' + lib + '_1.unpaired.trimmed.fastq.gz ' + \
            lib + '_2.paired.trimmed.fastq.gz ' + lib + '_2.unpaired.trimmed.fastq.gz ' + trimCmdB + '\n'
            
HTML(textarea('Trimmomatic bash commands', trimmomaticBash))
```

### Trimmomatic output

```python
#load trimming log file (trimm.err) and check
trimmLib = {}
prev = ''
trimmLog = open('/home/share/projects/smed_strain/trimm/trimm.err')
for line in trimmLog:
    if line.find('Started with arguments') != -1:
        prev = line.strip().split('/fastq/')[1].split('_1.fastq.gz')[0]
    if line.find('Surviving') != -1:
        meta = line.strip().split(':')
        numReads = int(meta[1].split()[0])
        keys = ['Reads'] + [meta[1].split()[-1]] + [' '.join(x.split()[2:]) for x in meta[2:-1]]
        vals = [(numReads,100)] + [(int(x.split()[:2][0]),float(x.split()[:2][1][1:-2])) for x in meta[2:]]
        kv = dict(zip(keys,vals))
        trimmLib[prev] = kv

labMeta = []
for lib, meta in trimmLib.items():
    labMeta.append((illum[lib], meta['Reads'][0], meta['Dropped'][1], meta['Surviving'][0]))
    
labMeta.sort(key = lambda x : x[0])
#filter out "Rajewsky asexual(SRR125365)". It only contains 50 reads. Wrongly uploaded file?
labMeta = filter(lambda x : x[0][0] != 'SRR125365', labMeta)

print 'Total reads (pairs):', commas(sum([x[1] for x in labMeta]))
print 'Asexual reads (pairs):', commas(sum([x[1] for x in labMeta if x[0][3] == 'asexual']))
print 'Sexual reads (pairs):', commas(sum([x[1] for x in labMeta if x[0][3] == 'sexual']))
print
print 'Total reads (pairs) after trimming:', commas(sum([x[3] for x in labMeta]))
print 'Asexual reads (pairs) after trimming:', commas(sum([x[3] for x in labMeta if x[0][3] == 'asexual']))
print 'Sexual reads (pairs) after trimming:', commas(sum([x[3] for x in labMeta if x[0][3] == 'sexual']))
print
for lab in list(set([x[0][2] for x in labMeta])):
    print 'Total ' + lab + ' reads (pairs):', commas(sum([x[3] for x in labMeta if x[0][2] == lab]))
trimmTable = ListTable()
trimmTable.append(['Lab','strain','Run','Platform','Total Reads (Pairs)','Surviving','Percentage Dropped','Study'])
trimmTable.extend([[x[0][2], x[0][3], x[0][0], x[0][9], commas(x[1]), commas(x[3]), str(x[2]),x[0][10]] for x in labMeta])
trimmTable
```

### Map reads with BWA

Trimmed PE and SE are mapped to the genome. Trimmed unpaired reads will be mapped as SE reads. 

```python
trimmedFs = [x for x in os.listdir('/home/share/projects/smed_strain/trimm') if x.split('.')[-1] == 'gz']
print 'Number of .fastq files:', len(trimmedFs)
```

The paired fastq files don't have the corrected formatted fastq headers. Wrote an interleave script to rename header and pipe to bwa as an interleaved file.

```python
#Interleave script
HTML(textarea('Interleave python script', '''import sys,gzip

pe1 = gzip.open(sys.argv[1],'rb')
pe2 = gzip.open(sys.argv[2],'rb')

for line in pe1:
        print '.'.join(line.split()[0].split('.')[:-1]) + "\\n" + \\
        pe1.next() + \\
        '.'.join(pe1.next().split()[0].split('.')[:-1]) + '\\n' + \\
        pe1.next() + \\
        '.'.join(pe2.next().split()[0].split('.')[:-1]) + '\\n' + \\
        pe2.next() + \\
        '.'.join(pe2.next().split()[0].split('.')[:-1]) + '\\n' + \\
        pe2.next().strip()
'''))
```

```python
prePath = '/home/share/projects/smed_strain/trimm/'
pecmdA = 'python interleave.py '  
pecmdB = ' | bwa mem -t 32 -p -R "'
pecmdC = '" /home/share/genome/smed/bwa/smed3.1 - | samtools view -hSbu - | samtools sort - ' 

secmdA = 'bwa mem -t 32 -R "'
secmdB = '" /home/share/genome/smed/bwa/smed3.1 '
secmdC = ' | samtools view -hSbu - | samtools sort - '

oenalCmdA = 'bwa mem -t 32 -R "'
oenalCmdB = '" /home/share/genome/smed/bwa/smed3.1 '
oenalCmdC = ' | samtools view -hSbu - | samtools sort - '

mappingCmds = ''
for item in labMeta:
    run = item[0][0]
    fastqs = [x for x in trimmedFs if x.split('_')[0] == run]
    rg = '@RG\\tID:' + run + '\\tlab:' + item[0][2] + '\\tstrain:' + item[0][3]
    if run.find('oenal') == -1:
        if len(fastqs) == 1:
            #SE 
            se = prePath + fastqs[0]
            mappingCmds += secmdA + rg + secmdB + se + secmdC + run + '.se.sorted\n'
        else:
            #PE
            pe1 = prePath + run + '_1.paired.trimmed.fastq.gz'
            pe2 = prePath + run + '_2.paired.trimmed.fastq.gz'
            mappingCmds += pecmdA + pe1 + ' ' + pe2 + pecmdB + rg + pecmdC + run + '.pe.sorted\n'

            #map unpaired as SE
            se1 = prePath + run + '_1.unpaired.trimmed.fastq.gz'
            se2 = prePath + run + '_2.unpaired.trimmed.fastq.gz'
            mappingCmds += secmdA + rg + secmdB + se1 + secmdC + run + '.unpe1.sorted\n'
            mappingCmds += secmdA + rg + secmdB + se2 + secmdC + run + '.unpe2.sorted\n'
    else:
        if len(fastqs) == 1:
            #SE 
            se = prePath + fastqs[0]
            mappingCmds += secmdA + rg + secmdB + se + secmdC + run + '.se.sorted\n'
        else:
            #PE
            pe1 = prePath + run + '_1.paired.trimmed.fastq.gz'
            pe2 = prePath + run + '_2.paired.trimmed.fastq.gz'
            mappingCmds += oenalCmdA + rg + oenalCmdB + pe1 + ' ' + pe2 + oenalCmdC + run + '.pe.sorted\n'

            #map unpaired as SE
            se1 = prePath + run + '_1.unpaired.trimmed.fastq.gz'
            se2 = prePath + run + '_2.unpaired.trimmed.fastq.gz'
            mappingCmds +=  secmdA + rg + secmdB + se1 + secmdC + run + '.unpe1.sorted\n'
            mappingCmds +=  secmdA + rg + secmdB + se2 + secmdC + run + '.unpe2.sorted\n'
            
HTML(textarea('BWA Mapping bash commands', mappingCmds))
```

### Categorize and merge resulting .BAM files a common sample files

Samtools merge is finicky to use. Instead, I am creating a separate `.sam` file for each grouping (lab.sex) with the correct @RG line and appending the corresponding `.bam` files to the correct header `.sam` file. This can all be done on bash. 

```python
bams = []
for item in labMeta:
    bams.append([item[0][0], item[0][2] + '.' + item[0][3]])
    
bams.sort(key = lambda x : x[1])

bamFiles = [x for x in os.listdir('/home/share/projects/smed_strain/bwa_map') if x.split('.')[-1] == 'bam']

from collections import defaultdict

categories = defaultdict(list)
for bam in bams:
    if bam[1].find('None') == -1:
        categories[bam[1]].append(bam[0])

bwa_categorizeCmds = ''
print 'Samples:'
for k,v in categories.items():
    print '    ' + k
    meta = k.split('.')
    #copy the common @SQ lines
    cmd = 'cp /home/share/projects/smed_strain/header /home/share/projects/smed_strain/' + k + '.sam'
    
    for run in v:
        #add specific @RG line for each run
        rg = '@RG\\tID:' + run + '\\tSM:' + k
        rgcmd = 'echo "' + rg + '" >> /home/share/projects/smed_strain/' + k + '.sam'
        cmd += '\n' + rgcmd
    for run in v:
        for bamFile in bamFiles:
            if bamFile.find(run) != -1:
                #append bam file to the lab.strain.sam file
                appendcmd = 'samtools view -F 4 /home/share/projects/smed_strain/bwa_map/' + bamFile + ' >> /home/share/projects/smed_strain/' + k + '.sam'
                cmd += '\n' + appendcmd
    bwa_categorizeCmds += cmd + '\n'

print '    Aboobaker.asexual'
HTML(textarea('Categorize BWA mapping output bash commands',bwa_categorizeCmds))
```

```python
#For ABI SOLiD mapping files
#RG:Z:ERR032072
solidFs = [x for x in os.listdir('/home/share/projects/smed_strain/novo_map/done')]
solidFs = [x for x in solidFs if x.split('.')[-1] == 'sam']

novoalign_categorizeCmds = 'samtools view -SH /home/share/projects/smed_strain/novo_map/done/' + solidFs[0] + ' > /home/share/projects/smed_strain/Aboobaker.asexual.sam'
for f in solidFs:
    rg = '@RG\\tID:' + f.split('.')[0] + '\\tSM:Aboobaker.asexual'
    rgcmd = 'echo "' + rg + '" >> /home/share/projects/smed_strain/Aboobaker.asexual.sam'
    novoalign_categorizeCmds += rgcmd + '\n'

for f in solidFs:
    novoalign_categorizeCmds += 'samtools view -S -F 4 /home/share/projects/smed_strain/novo_map/done/' + f + " | sed -e 's/$/\\tRG:Z:" + f.split('.')[0] + "/' >> /home/share/projects/smed_strain/Aboobaker.asexual.sam\n"
    
HTML(textarea('Categorize novoalign output bash commands',novoalign_categorizeCmds))
```

### Post processing of mapped reads

Convert and sort each sample to .bam. Use picard tools to remove duplicates. 

```python
consolidated = [x for x in os.listdir('/home/share/projects/smed_strain/consolidated_map/raw') if x.split('.')[-1] == 'sam']

sortCmds = ''
for conFile in consolidated:
    name = '.'.join(conFile.split('.')[:2])
    sortCmds += 'samtools view -bSh /home/share/projects/smed_strain/consolidated_map/raw' + conFile + ' | samtools sort - ' + name + '.sorted\n'

HTML(textarea('Sort consolidated .sam files bash commands',sortCmds))
```

```python
dedupCmds = ''

for conFile in consolidated:
    name = '.'.join(conFile.split('.')[:2])
    dedupCmds += '''java -jar /home/share/software/picard-tools-1.115/MarkDuplicates.jar MAX_FILE_HANDLES_FOR_READ_ENDS_MAP=60 INPUT=''' + name + '''.sorted.bam OUTPUT=''' + name + '''.sorted.dedup.bam METRICS_FILE=''' + name + '''.metrics.txt\n'''
    
HTML(textarea('Piccard-tools de-duplicate bash commands',dedupCmds))
```

```python
indexCmds = ''

for conFile in consolidated:
    name = '.'.join(conFile.split('.')[:2])
    indexCmds += '''java -jar /home/share/software/picard-tools-1.115/BuildBamIndex.jar INPUT=''' + name + '''.sorted.dedup.bam\n'''
    
HTML(textarea('Index sorted, deduped bash commands',indexCmds))
```
